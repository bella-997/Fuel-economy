### CLEANING PROCESS

```python
df1 = df_08.copy() #a copy of 2008 dataset
df2 = df_18.copy() #a copy of 2018 dataset
```

```python
#droping columns in 2008
L = ['Stnd', 'Underhood ID', 'FE Calc Appr', 'Unadj Cmb MPG','Sales Area']
df1.drop(L, axis = 1, inplace = True)
df1.columns
```

```python
#drop columns in 2018
M = ['Stnd', 'Stnd Description', 'Underhood ID', 'Comb CO2','Cert Region']
df2.drop( M, axis = 1, inplace = True)
df2.columns
```

```python
# replace spaces with underscores and lowercase labels for 2008 dataset
df1.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)

# replace spaces with underscores and lowercase labels for 2008 dataset
df2.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)
# confirm changes
(df1.columns == df2.columns).all() #checking if the columns match for both datasets
```

```python
#check for duplicated rows in 2008
df1.duplicated().sum()
```

```python
#checking for duplicated rows in 2018
df2.duplicated().sum()
```

```python
#dropping duplicated rows in both years
df1.drop_duplicates(inplace = True)
df2.drop_duplicates(inplace = True)

print(df1.duplicated().sum()) #checking for duplicates
print(df2.duplicated().sum())
```

```python
df1.isnull().sum() #checking again for null values
```

101 missing values doesn't look good. I can't replace them because it will be too much of guess data, so i have to drop them. Drive column has 40 missing values and its part of the rows that have missing values.

```python
#looking at the drive column
df1[df1['cyl'].isnull()]
```

```python
#dropping missing values
df1.dropna(axis = 0, inplace = True)
df1.isnull().sum().any() #checking for missing values
```

```python
df2.isna().sum() #checking for missing values in 2018
```

there are 2, one in displ and the other in cyl, so i'll fill them up with the mean values

```python
df2[df2['displ'].isnull()]
```

```python
#replacing the null values with the mean of the cyl and displ
df2['displ'].fillna(df2['displ'].mean(), inplace = True)
df2['cyl'].fillna(df2['cyl'].mean(), inplace = True)

df2.isna().sum().any()
```

##### converting the cyl column in 2008 from str to int

```python
df1['cyl'] = df1['cyl'].str.extract('(\d+)').astype(int)
df1['cyl'].value_counts()
```

```python
#converting cyl in 2018 from float to int dtype in cyl 2018
df2['cyl'] = df2['cyl'].astype(int)
df2['cyl'].dtype
```

### Fixing hybrid values in fuel, air_pollution_score, the mpgs and greenhouse_gas_score

```python
#first we have to check the fuel types that contain 2 values

df1[df1['fuel'].str.contains('/')] #2008 dataset
```

```python
df2[df2['fuel'].str.contains('/')]
```

For the cleaning of this part i've decided to leave the `fuel` column, since it's a string and will remain as such, but for other columns which are meant to be numerical and are affected by this hybrid fuel types, I'll take their average so as to keep the columns analyzable. so first I'll split the values, add them up, fine the mean and convert to int. Tadaaa

```python
hbd = ['air_pollution_score','city_mpg','hwy_mpg','cmb_mpg','greenhouse_gas_score'] #hybrid columns
for i in hbd:
    print((df1[i].dtype))
```

```python
df1['air_pollution_score'].unique()
```

```python
#splitting the 2008 hybrid columns
#split the values and calculate the average
for i in hbd:
    df1[i] = df1[i].str.split('/').apply(lambda x: int((int(x[0]) + int(x[1]))/2) if len(x) == 2 else (x[0]))

df1[df1['fuel'].str.contains('/')]
```

```python
#splitting the 2018 hybrid columns
hbd2 = ['city_mpg','hwy_mpg','cmb_mpg'] #hybrid columns

#split the values and calculate the average
for i in hbd2:
    df2[i] = df2[i].str.split('/').apply(lambda x: int((int(x[0]) + int(x[1]))/2) if len(x) == 2 else (x[0]))

df2[df2['fuel'].str.contains('/')]
```
